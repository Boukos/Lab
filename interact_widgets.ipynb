## Intro to `@interact`

Note: This notebook is based partly on the notebook here: https://github.com/ipython/ipywidgets/blob/master/docs/source/examples/Using%20Interact.ipynb

```python
from ipywidgets import interact, interactive, fixed
import ipywidgets as widgets
```

At the most basic level, interact autogenerates UI controls for function arguments, and then calls the function with those arguments when you manipulate the controls interactively. To use interact, you need to define a function that you want to explore. Here is a function that prints its only argument x.

```python
def f(x):
    return x
```

```python
# ints make sliders
interact(f, x=10.0)
```

```python
# bools make checkboxes
interact(f, x=False)
```

### Plotting Example

```python
import quandl
import matplotlib.pyplot as plt
%matplotlib inline
```

Let's grab some data from quandl. We'll work with the database `WIKI`, which contains end of day data on stock prices. 

```python
def plot_stocks(ticker, column):
    df = quandl.get("WIKI/"+ticker, start_date="2016-01-01")
    ax = df[column].plot()
    ax.set_title("{} Stock data for {}".format(column, ticker))
```

```python
interact(plot_stocks, ticker="FB", column=["Open", "High", "Low", "Close"])
```

There is also a syntax that uses `interact` as a decorator. The example above would look like this:

```python
@interact(ticker="FB", column=["Open", "High", "Low", "Close"])
def plot_stocks(ticker, column):
    df = quandl.get("WIKI/"+ticker, start_date="2016-01-01")
    ax = df[column].plot()
    ax.set_title("{} Stock data for {}".format(column, ticker))
```
